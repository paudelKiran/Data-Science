# Introduction to scikit learn
Intro to some functions of scikit learn

### 1. Getting Data Ready

```python
import pandas as pd
import numpy as np
heartDisease=pd.read_csv("heart-disease.csv")
```

```python
heartDisease
```

```python
# x as Feature matrix
x=heartDisease.drop("target",axis=1)

# y as labels (data that doesn't to be manipulated or most of the time output)
y=heartDisease['target']
x,y
```

### 2.Choosing the right model

```python
import sklearn
print(sklearn.__version__)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
# keeping default hyper parameters
clf.get_params()
```

## 3. Fitting model to training data

```python
from sklearn.model_selection import train_test_split

x_train ,x_test,y_train, y_test=train_test_split(x,y,test_size=0.2)
```

```python
clf.fit(x_train,y_train) #giving instruction to clf to find pattern in our data
```

```python
#making predictions
len(x)
```

```python
x_test
```

```python
ypreds=clf.predict(x_test) #test data ma predict garna lagaiyo
ypreds
```

## Evaluating a model

```python
clf.score(x_train,y_train) #training data
```

```python
clf.score(x_test,y_test)
```

```python
# Other ways of evaluating a model
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print(classification_report(y_test,ypreds))
```

```python
confusion_matrix(y_test,ypreds)
```

## Improving a model


```python
#trying different amount of n_estimators
for i in range(10,101,10):
  clf=RandomForestClassifier(n_estimators=i,random_state=42)
  clf.fit(x_train,y_train)
  print(f"Model accuracy= {clf.score(x_test,y_test)} when i={i}")
```
