### Working with pandas

```python
carSales=pd.read_csv("car-sales.csv")
```

```python
carSales
```

```python
carSales['Price']=carSales['Price'].str.replace(r"[\$\,\.]",'',regex=True)
carSales['Price']=carSales['Price'].str[:-2].astype(int)
carSales 
```

```python
carSales['Price'].dtype
```

```python
#adding sales date
carSales["Sales Date"]=pd.date_range("1/1/2025",periods=len(carSales))
carSales
```

```python
carSales["Total Sales Amount"]=carSales['Price'].cumsum() #calculating cumulative sum
carSales
```

```python
carSales.plot(y='Total Sales Amount',x='Sales Date',color="green")
```
